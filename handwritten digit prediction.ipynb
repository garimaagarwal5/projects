```python

(x_train,y_train),(x_test,y_test)=tf.keras.datasets.mnist.load_data()#mist is dataset having many images for handwritten digits
```

```python
x_train[0]
```

```python
import matplotlib.pyplot as plt
plt.imshow(x_train[7],cmap='gray')#imap converts pixels into ,0 stands for black color,1 stands for white color
plt.show()
```

```python
#normalization is the process when our values are dur dur to we deal that case with normalization
```

```python
x_train=x_train/255.0
```

```python
x_test=x_test/255.0
```

```python
x_train[0]
```

```python
x_train=x_train.reshape(-1,28,28,1)#reshape in 28 rows and 28 columns,-1 automatically indicate how many images are in set,a indicate to change 2d into 3 d
x_test=x_test.reshape(-1,28,28,1)
```

```python
x_train[0]
```

```python
y_train=tf.keras.utils.to_categorical(y_train,num_classes=10)#num_classes are no. of digits in o-9,10 unique values
y_test=tf.keras.utils.to_categorical(y_test,num_classes=10)
```

```python
model=tf.keras.Sequential()#makes a container to build neural network
model.add(tf.keras.layers.Conv2D(32,(3,3),activation='relu',input_shape=(28,28,1)))#creates a layer in form of ,1st hideen layer,32 filters are there,kernel is filter
#that rotate on imnage,ativation funtion is relu,image shape is 28 rows and 28 columns,kernel size is 3x3

model.add(tf.keras.layers.MaxPooling2D(2,2))#after extrctiong feature mapping we are doing maximum pooling
model.add(tf.keras.layers.Flatten())#makes output into flatten (form of 1D) 
model.add(tf.keras.layers.Dense(128,activation='relu'))
model.add(tf.keras.layers.Dense(10,activation='softmax'))#gives 10 output i.e what is probabitlty of 0 or 1 or 2
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])#optimizer to reduce errors.
```

```python
model.fit(x_train,y_train,epochs=10,batch_size=128,validation_split=0.2)
```

```python
def sketch_recognition(img):
    x=model.predict(img.reshape(1,28,28)).argmax(axis=1)
    return x
```

```python
accuracy=model.evaluate(x_test,y_test)
accuracy
```
